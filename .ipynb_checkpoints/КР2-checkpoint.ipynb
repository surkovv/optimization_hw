```python
import numpy as np
import pandas as pd
import scipy.stats as sps
import scipy.optimize as optimize
import matplotlib.pyplot as plt
import seaborn as sns
```

Возьмём реализацию CALMNS-LP из статьи

```python
def CAMNS_LP(C, d, L, N=3):
    
    # Step 1
    l = 0
    B = np.eye(L)
    
    while l != N:
    
        # Step 2
        w = sps.norm().rvs(L)
        
        r = B.dot(w)
        
        # Step 3

        def fun(alpha):
            return r.dot(C @ alpha + d.flatten())
        
        constraint = optimize.LinearConstraint(C, lb=-d.flatten(), ub=np.ones(C.shape[0]) * np.inf)

        alpha1 = optimize.minimize(fun, 
                                   x0=np.zeros(N-1),
                                   method='COBYLA',
                                   constraints=constraint).x

        alpha2 = optimize.minimize(lambda x: -fun(x), 
                                   x0=np.zeros(N-1),
                                   method='COBYLA',
                                   constraints=constraint).x
        p_star = fun(alpha1)
        q_star = fun(alpha2)

        # Step 4
        
        if l == 0:
            S = np.vstack((C @ alpha1 + d.T, C @ alpha2 + d.T)).T
        else:
            if np.linalg.norm(p_star) != 0:
                column = C @ alpha1 + d.T
                S = np.hstack((S, column.reshape(-1, 1)))
            elif np.linalg.norm(q_star) != 0:
                column = C @ alpha2 + d.T
                S = np.hstack((S, column.reshape(-1, 1)))

        # Step 5
        l = S.shape[1]

        print(l)
        # Step 6

        q, r = np.linalg.qr(S)
        B = np.eye(L) - q @ q.T
    
    return S
    
```

Нахождение матриц $C$ и столбца $d$, задающих афинное множество, также возьмем из статьи.

```python
def find_C_d(X, N=3):
    d = X.mean(axis=1).reshape(-1, 1)
    U = X - np.tile(d, (1, X.shape[1]))
    C = U[:, :N-1]
    return C, d
```

Проверим, что все работает на банальном примере

```python
X_start = np.array([
    [1, 0, 0],
    [0, 1, 0],
    [0, 0, 1]
])

A = np.array([
    [0.4, 0.7, 0.1],
    [0.3, 0.1, 0.1],
    [0.3, 0.2, 0.8]
])

X = X_start @ A
C, d = find_C_d(X)
CAMNS_LP(C, d, L=3, N=3)
```

Теперь перейдем к картинкам. Возьму фотографии и смешаю (пропорции стоят в столбцах).

```python
names = ['pics/cao1.jpg', 'pics/ksiwek1.jpg', 'pics/zhang1.jpg']
imgs = np.array([plt.imread(name) for name in names])

imgs_flat = np.array([img.flatten() for img in imgs])

weights = np.array([
    [0.4, 0.6, 0.1],
    [0.5, 0.3, 0.2],
    [0.1, 0.1, 0.7]
])

mix = imgs_flat.T @ weights
```

```python
plt.figure(figsize=(15, 10))
for i in range(3):
    ax = plt.subplot(1, 3, i + 1)
    ax.imshow(mix[:, i].reshape(128, 128), cmap='gray')
plt.show()
```

Запустим метод.

```python
C, d = find_C_d(mix)
result = CAMNS_LP(C, d, mix.shape[0])
```

Вот что получилось

```python
plt.figure(figsize=(15, 10))
for i in range(3):
    ax = plt.subplot(1, 3, i + 1)
    ax.imshow(result[:, i].reshape(128, 128), cmap='gray')
plt.show()
```

Посмотрим, насколько метод хорошо работает на похожих объектах. В данном случае, фотографии одного человека с разными выражениями лица.

```python
names = ['pics/zhenya1.jpg', 'pics/zhenya2.jpg', 'pics/zhenya3.jpg', 'pics/zhenya4.jpg']
imgs = np.array([plt.imread(name) for name in names])


imgs_flat = np.array([img.flatten() for img in imgs])

weights = np.array([
    [0.4, 0.6, 0.1, 0.05],
    [0.3, 0.1, 0.1, 0.1],
    [0.2, 0.1, 0.7, 0.05],
    [0.1, 0.2, 0.1, 0.8]
])

mix = imgs_flat.T @ weights
```

```python
plt.figure(figsize=(15, 10))
for i in range(4):
    ax = plt.subplot(1, 4, i + 1)
    ax.imshow(mix[:, i].reshape(*imgs[i].shape) / 255)
plt.show()
```

```python
C, d = find_C_d(mix, N=4)
result = CAMNS_LP(C, d, mix.shape[0], N=4)
plt.figure(figsize=(15, 10))
for i in range(4):
    ax = plt.subplot(1, 4, i + 1)
    res = np.maximum(result[:, i] / 255, 0)
    res = np.minimum(res, 1)
    ax.imshow(res.reshape(*imgs[i].shape))
plt.show()
```

Сравним с исходными фото.

```python
plt.figure(figsize=(15, 10))
for i in range(4):
    ax = plt.subplot(1, 4, i + 1)
    ax.imshow(imgs[i])
plt.show()
```
